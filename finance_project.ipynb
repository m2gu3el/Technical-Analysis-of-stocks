```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=yf.download('AAPL', start='2017-01-01', period="1d" )
symbol="AAPL"
```

```python
df
```

```python
# Create a chart for the closing price versus date
plt.plot(df['Close'])
plt.title(f'{symbol} Closing Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
df['MA20'] = df['Adj Close'].rolling(20).mean()
df['MA50'] = df['Adj Close'].rolling(50).mean()
```

```python
df['MACD'] = df['MA20'] - df['MA50']
df['Signal_Line'] = df['MACD'].rolling(10).mean()
```

```python
df
```

```python
df=df.dropna()
```

```python
df
```

```python
buy = []
sell = []
position = 0
position_list = [0]
stop_loss_pct = 0.1
take_profit_pct = 0.1

for i in range(1, len(df)):
    if df['MACD'].iloc[i] > df['Signal_Line'].iloc[i] and df['MACD'].iloc[i - 1] <= df['Signal_Line'].iloc[i - 1]:
        buy.append(i)
        if position == 0:
            position = 1
            entry_price = df['Close'].iloc[i]
        position_list.append(position)
    elif df['MACD'].iloc[i] < df['Signal_Line'].iloc[i] and df['MACD'].iloc[i - 1] >= df['Signal_Line'].iloc[i - 1]:
        sell.append(i)
        if position == 0:
            position = -1
            entry_price = df['Close'].iloc[i]
        elif position == 1:
            exit_price = df['Close'].iloc[i]
            stop_loss_price = entry_price * (1 - stop_loss_pct)
            if df['Low'].iloc[i] < stop_loss_price:
                exit_price = stop_loss_price
            elif df['High'].iloc[i] > entry_price * (1 + take_profit_pct):
                exit_price = entry_price * (1 + take_profit_pct)
            position = 0
        position_list.append(position)

    else:
        position_list.append(position)








```

```python
df['position']=position_list
```

```python
max_drawdown = 0
peak_portfolio_value = df['Portfolio Value'].iloc[0]

for i in range(1, len(df)):
    current_value = df['Portfolio Value'].iloc[i]

    if current_value > peak_portfolio_value:
        peak_portfolio_value = current_value
    else:
        drawdown = (current_value - peak_portfolio_value) / peak_portfolio_value * 100
        max_drawdown = min(max_drawdown, drawdown)

print("Maximum Drawdown (%):", max_drawdown)
```

```python
class MyMACDStrategy(Strategy):
  def init(self):
    price=self.data.Close
    self.macd=self.I(lambda x: talib.MACD(x)[0], price)
    self.macd_signal=self.I(lambda x: talib.MACD(x)[1], price)

  def next(self):
    if crossover(self.macd, self.macd_signal):
      self.buy()
    elif crossover(self.macd_signal, self.macd):
      self.sell()

```

```python

start = dt.datetime(2017, 1, 1)
end = dt.datetime(2023, 1, 1)

ticker = yf.Ticker("AAPL")

data = ticker.history(start=start, end=end)
```

```python
backtest= Backtest(data, MyMACDStrategy, commission=0.002, exclusive_orders=True)
print(backtest.run())
```
